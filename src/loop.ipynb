```python
import os
from ultralytics import YOLO
from loop.main_loop import run_tracking

video_or_frames_path = r""
output_dir = r""
model_path = r""

classes = {
    0: {"color": (255, 255, 0), "tag": "Multicopter"},
    1: {"color": (255, 0, 255), "tag": "Wing"},
    2: {"color": (0, 255, 255), "tag": "XWing"}
}

if os.path.exists(video_or_frames_path):
    model = YOLO(model_path)

    if os.path.isdir(video_or_frames_path):
        print(f"Processing frames from folder: {video_or_frames_path}")
        run_tracking(
            model=model,
            frames_dir=video_or_frames_path,
            output_dir=output_dir,
            visualize=True,
            classes=classes
        )
    else:
        print(f"Processing video file: {video_or_frames_path}")
        run_tracking(
            model=model,
            video=video_or_frames_path,
            output_dir=output_dir,
            visualize=True,
            classes=classes
        )
else:
    print("File or folder not found")
```
